```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import pandas as pd
```

```python
def plot_training(df, mode, y_axis, title):
    
    a = 'train_'+mode+'_mean'
    b = 'train_'+mode+'_std'
    fig = go.Figure([
        go.Scatter(
            name='Train-Accuracy',
            x=df.index,
            y=df[a],
            mode='lines',
            line=dict(color='rgb(31, 119, 180)'),
        ),
        go.Scatter(
            name='Upper Bound',
            x=df.index,
            y=df[a]+df[b],
            mode='lines',
            marker=dict(color="#444"),
            line=dict(width=0),
            showlegend=False
        ),
        go.Scatter(
            name='Lower Bound',
            x=df.index,
            y=df[a]-df[b],
            marker=dict(color="#444"),
            line=dict(width=0),
            mode='lines',
            fillcolor='rgba(68, 68, 68, 0.3)',
            fill='tonexty',
            showlegend=False
         ),
        go.Scatter(
            name='Valid-Accuracy',
            x=df.index,
            y=df['valid_acc_mean'],
            mode='markers',
            marker=dict(color='red', size=2),
        ),
        go.Scatter(
            name='Upper Bound',
            x=df.index,
            y=df['valid_acc_mean']+df['valid_acc_std'],
            mode='lines',
            marker=dict(color="#444"),
            line=dict(width=0),
            showlegend=False
        ),
        go.Scatter(
            name='Lower Bound',
            x=df.index,
            y=df['valid_acc_mean']-df['valid_acc_std'],
            marker=dict(color="#444"),
            line=dict(width=0),
            mode='lines',
            fillcolor='rgba(68, 68, 68, 0.3)',
            fill='tonexty',
            showlegend=False
        )
    ])
    fig.update_layout(
        yaxis_title=y_axis,
        title=title,
        hovermode="x"
    )
    fig.show()
```
